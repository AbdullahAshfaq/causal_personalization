```python
import causallearn
from causallearn.search.ConstraintBased.PC import pc
import pandas as pd
import numpy as np
```

```python
# Loading data
file_url = "https://msalicedatapublic.z5.web.core.windows.net/datasets/ROI/multi_attribution_sample.csv"
multi_data = pd.read_csv(file_url)
multi_data.head()
```

```python
cg = pc(multi_data.to_numpy())
```

```python
cg
```

```python
cg.draw_pydot_graph(labels=multi_data.columns)
```

## FCI 

```python
from causallearn.search.ConstraintBased.FCI import fci

cg_fci, edges = fci(multi_data.to_numpy())
```

```python
cg_fci.draw
```

```python
from causallearn.utils.GraphUtils import GraphUtils

pyd = GraphUtils.to_pydot(cg_fci, labels=multi_data.columns)
```

```python
# To display pydot graph
from IPython.display import Image, display

def view_pydot(pdot):
    plt = Image(pdot.create_png())
    display(plt)
```

```python
view_pydot(pyd)
```
